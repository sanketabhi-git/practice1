```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
cap = cv2.VideoCapture('/content/robots.mp4')

while cap.isOpened():
  ret, frame = cap.read()
  if not ret:
    print('End of the video file...')
    break
  frame_rgb=cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
  results=model(frame_rgb)
  results.render()

  final_frame=cv2.cvtColor(results.ims[0], cv2.COLOR_RGB2BGR)
  cv2.imshow('yolov5 Object Detection;',final_frame)
  if cv2.waitKey(1)&0xFF==ord('q'):
    break

cap.release()
cv2.destroyAllWindows()
```
