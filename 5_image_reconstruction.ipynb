```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('images/cat_damaged.png')
plt.imshow(image)
plt.show()



```

```python
height, width = image.shape[:-1]
```

```python
height
```

```python
width
```

```python
cv2.imshow('Damaged cat', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
mask = np.zeros([height, width, 3], dtype=np.uint8)
```

```python
# Converting all pixels greater than zero to black while black becomes white

for i in range(height):
    for j in range(width):
        if image[i, j].sum() == 0:
            mask[i, j] = [255, 255, 255]
```

```python
plt.subplot(1,2,1)
plt.imshow(image)
plt.subplot(1,2,2)
plt.imshow(mask, cmap='gray')
```

```python
mask = cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)
```

```python
# Inpaint.
dst = cv2.inpaint(image, mask, 3, cv2.INPAINT_NS)
```

```python
cv2.imshow('hi',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
