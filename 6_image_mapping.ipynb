```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/dog.jpg')
plt.imshow(img)
plt.show()
```

```python
rows,  cols  =  img.shape[:2]
```

```python
rows
```

```python
cols
```

```python
# create the grid
gridx, gridy = np.meshgrid(np.arange(0, cols), np.arange(0, rows))
```

```python
gridy
```

```python
# define the mapping rule
mapx  =  gridx
mapy = gridy+ 15 + np.sin(gridx/20.)*10
```

```python
plt. plot(np.arange(1, 40, 0.1), np.sin(np.arange(1, 40, 0.1)))
```

```python
mapx = mapx.astype(np.float32)
mapy = mapy.astype(np.float32)
```

```python
mapped_img = cv2.remap(img, mapx, mapy, interpolation=cv2.INTER_LINEAR)
```

```python
mapped_img = cv2.cvtColor(mapped_img, cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(mapped_img)
```
